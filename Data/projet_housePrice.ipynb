# <font color ="orange">>> Création du Modele


Dans cette partie on va appliquer plusieurs types d'algorithme de regression et choisir celui qui donne plus de performance

```python
# séparer les attributs du label 'saleprice'
y = df_train['SalePrice']
X = df_train.drop('SalePrice',axis=1)
```

Division du jeux de données en données d'entrainement et données de test


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=324)
```

## <font color ="green"> 1. K_Nearest Neighbors

tourner le modele sur plusieurs valeurs de k et calculer l'erreur produit pour chaque valeur

```python
from sklearn.neighbors import KNeighborsRegressor
from math import sqrt

from sklearn.metrics import mean_squared_error
for j in range(1,21):
    neigh = KNeighborsRegressor(n_neighbors=j)
    neigh.fit(X_train, y_train) 
    y_pKNN=neigh.predict(X_test)
    #print(y_pKNN[:5])
    #y_test[:5]
    
    RMSEKNN1 = sqrt(mean_squared_error(y_true = np.log(y_test), y_pred = np.log(y_pKNN)))
    print("RMSLE POUR K = {} KNN: {}".format(j,RMSEKNN1))
## for
```

<li style="color:red"> On va donc choisir la  valeur de K ayant la plus petite valeur de RMSLE, K=10

Effectuer notre prédiction 

```python
neigh = KNeighborsRegressor(n_neighbors=10)
neigh.fit(X_train, y_train) 
y_pKNN=neigh.predict(X_test)
print("les valeurs prédites : ",y_pKNN)
print("les valeurs réelles : ",y_test)
```

Calculer le score d'accuracy de la prédiction effectuée avec l'algorithme KNN 

```python
scoreKNN=neigh.score(X_test, y_test)
print("le score est :{} ".format(scoreKNN)*100)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

RMSEKNN = sqrt(mean_squared_error(y_true = y_test, y_pred = y_pKNN))
print("RMSE POUR KNN : {}".format(RMSEKNN))

RMSEKNN1 = sqrt(mean_squared_error(y_true = np.log(y_test), y_pred = np.log(y_pKNN)))
print("RMSLE POUR KNN: {}".format(RMSEKNN1))

MAE=mean_absolute_error(y_test, y_pKNN)
print("MAE POUR KNN: {}".format(MAE))
```

## <font color="blue">2. Appliquer flask 

le fichier index1.html est l'interface de notre application contenat un formulaire, donc il suffit de l'appeler et appeler le modele crée qu'on a deja enregistrer

```python
import numpy as np
from flask import Flask, request, jsonify, render_template
import pickle

app = Flask(__name__)
model = pickle.load(open('model.pkl', 'rb'))

@app.route('/')
def home():
    return render_template('index1.html')

@app.route('/predict',methods=['POST'])
def predict():

    int_features = [int(x) for x in request.form.values()]
    final_features = [np.array(int_features)]
    prediction = model.predict(final_features)

    output = round(prediction[0], 2)

    return render_template('index1.html', prediction_text='Sales should be $ {}'.format(output))

@app.route('/results',methods=['POST'])
def results():

    data = request.get_json(force=True)
    prediction = model.predict([np.array(list(data.values()))])

    output = prediction[0]
    return jsonify(output)

if __name__ == "__main__":
    app.run(debug=True)
```
